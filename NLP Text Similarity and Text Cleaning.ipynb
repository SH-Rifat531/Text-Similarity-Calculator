```python
import tkinter as tk
import re
import string
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def clean_text(text):
    text = re.sub(r"n't", " not", text)
    text = re.sub(r"'d", " would", text)
    text = re.sub(r"'ll", " will", text)
    text = re.sub(r"'ve", " have", text)
    text = re.sub(r"'re", " are", text)
    text = re.sub(r"'m", " am", text)
    text = re.sub(r"'s", " is", text)
    text = re.sub(r"'ll", " will", text)
    
    text = re.sub(f'[{string.punctuation}]', '', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = text.lower()
    text = ' '.join(text.split())
    
    return text

def perform_text_cleaning():
    text_to_clean = input_text1.get("1.0", "end-1c")
    cleaned_text = clean_text(text_to_clean)
    cleaned_output.delete("1.0", tk.END)
    cleaned_output.insert(tk.END, cleaned_text)

def calculate_similarity():
    cleaned_text1 = cleaned_output.get("1.0", "end-1c")
    cleaned_text2 = input_text2.get("1.0", "end-1c")
    
    cleaned_text2 = clean_text(cleaned_text2)
    
    corpus = [cleaned_text1, cleaned_text2]
    
    vectorizer = CountVectorizer().fit_transform(corpus)
    vectors = vectorizer.toarray()
    
    similarity = cosine_similarity(vectors)[0][1]
    similarity_result.config(text=f"Similarity: {similarity:.2f}")

def handle_contractions():
    text_to_clean = input_text1.get("1.0", "end-1c")
    text_to_clean = re.sub(r"n't", " not", text_to_clean)
    text_to_clean = re.sub(r"'d", " would", text_to_clean)
    text_to_clean = re.sub(r"'ll", " will", text_to_clean)
    text_to_clean = re.sub(r"'ve", " have", text_to_clean)
    text_to_clean = re.sub(r"'re", " are", text_to_clean)
    text_to_clean = re.sub(r"'m", " am", text_to_clean)
    text_to_clean = re.sub(r"'s", " is", text_to_clean)
    text_to_clean = re.sub(r"'ll", " will", text_to_clean)
    
    cleaned_output.delete("1.0", tk.END)
    cleaned_output.insert(tk.END, text_to_clean)

def remove_punctuation():
    text_to_clean = input_text1.get("1.0", "end-1c")
    text_to_clean = re.sub(f'[{string.punctuation}]', '', text_to_clean)
    cleaned_output.delete("1.0", tk.END)
    cleaned_output.insert(tk.END, text_to_clean)

def remove_special_characters():
    text_to_clean = input_text1.get("1.0", "end-1c")
    text_to_clean = re.sub('\W+','', text_to_clean)
    cleaned_output.delete("1.0", tk.END)
    cleaned_output.insert(tk.END, text_to_clean)

def lowercase_text():
    text_to_clean = input_text1.get("1.0", "end-1c")
    text_to_clean = text_to_clean.lower()
    cleaned_output.delete("1.0", tk.END)
    cleaned_output.insert(tk.END, text_to_clean)

def eliminate_extra_spaces():
    text_to_clean = input_text1.get("1.0", "end-1c")
    text_to_clean = ' '.join(text_to_clean.split())
    cleaned_output.delete("1.0", tk.END)
    cleaned_output.insert(tk.END, text_to_clean)

root = tk.Tk()
root.title("Text Similarity Calculator")

input_frame1 = tk.Frame(root)
input_frame1.pack()

label1 = tk.Label(input_frame1, text="Enter Text 1:")
label1.pack()

input_text1 = tk.Text(input_frame1, height=5, width=40)
input_text1.pack()

cleaning_frame = tk.Frame(root)
cleaning_frame.pack()

contractions_button = tk.Button(cleaning_frame, text="Contractions", command=handle_contractions)
contractions_button.pack()

remove_punctuation_button = tk.Button(cleaning_frame, text="Remove Punctuation", command=remove_punctuation)
remove_punctuation_button.pack()

remove_special_chars_button = tk.Button(cleaning_frame, text="Remove Special Characters", command=remove_special_characters)
remove_special_chars_button.pack()

lowercasing_button = tk.Button(cleaning_frame, text="Lowercasing Text", command=lowercase_text)
lowercasing_button.pack()

eliminate_extra_spaces_button = tk.Button(cleaning_frame, text="Eliminate Extra Spaces", command=eliminate_extra_spaces)
eliminate_extra_spaces_button.pack()

cleaned_output = tk.Text(root, height=5, width=40)
cleaned_output.pack()

input_frame2 = tk.Frame(root)
input_frame2.pack()

label2 = tk.Label(input_frame2, text="Enter Text 2:")
label2.pack()

input_text2 = tk.Text(input_frame2, height=5, width=40)
input_text2.pack()

calculate_button = tk.Button(root, text="Calculate Similarity", command=calculate_similarity)
calculate_button.pack()

similarity_result = tk.Label(root, text="")
similarity_result.pack()

root.mainloop()
```
